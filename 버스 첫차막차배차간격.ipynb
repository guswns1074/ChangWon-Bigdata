```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from time import sleep
from bs4 import BeautifulSoup
import pandas as pd
from openpyxl import load_workbook

options = Options()
options.add_argument("--start-maximized")

 

def bus_interval(a):
    driver = webdriver.Chrome(executable_path="C:\\Users\CPB06GameN\Anaconda3\envs\chromedriver.exe", chrome_options=options)
    driver.get('http://bus.changwon.go.kr/search/route.do')
    soup = BeautifulSoup(driver.page_source, 'lxml')
    driver.find_element_by_css_selector("#tbodyList > tr:nth-child("+str(a)+") > td > a").click()
    time = driver.find_element_by_css_selector("#rouSearch > detail > div.timetable").text  
    interval = driver.find_element_by_css_selector("#rouSearch > detail > div.interval").text 
    driver.close()

    wb = load_workbook('버스배차.xlsx')
    sheet = wb.active
    sheet.cell(row=a, column=3).value = time
    sheet.cell(row=a, column=4).value = interval  
    wb.save('버스배차.xlsx')
```

```python
from datetime import datetime

new1 = df["첫차막차"].str.replace("평일","")
new2 = new1.str.replace("토요일","")
new3 = new2.str.replace("공휴일","")

interval = []

for i in range(0,len(new3)):
    first = new3[i][0:5]
    last = new3[i][7:12]
    interval.append(datetime.strptime(last,'%H:%M') - datetime.strptime(first,'%H:%M'))
```

```python
new4 = df["배차간격"].str.replace("평일","")
new5 = new4.str.replace("토요일","")
new6 = new5.str.replace("공휴일","")
new7 = new6.str.replace("분간격","")

interval2 = []

for i in range(0,len(new6)):
    a = new6[i].split("~")
    interval2.append((int(a[1].split("분")[0])+1+int(a[0]))/2)
```

```python
interval3 = []
for i in range(0,len(interval)):
    interval3.append(str(interval[i]))
```

```python
interval4 = []
for i in range(0,len(interval3)):
    interval4.append(int(interval3[i].split(":")[0])*60+int(interval3[i].split(":")[1]))
```

```python
interval_final = []
for i in range(0,len(interval4)):
    interval_final.append(interval4[i]/interval2[i])
```

```python
df["막차-첫차"]= interval3
```

```python
df["배차간 시간"] = interval2
```

```python
df["하루운행횟수"] = interval_final
```
