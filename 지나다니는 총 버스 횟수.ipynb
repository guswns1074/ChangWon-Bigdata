```python
import pandas as pd

bus_num_df = pd.read_excel("버스번호최종.xlsx")

bus_interval_df = pd.read_excel("배차간격최종.xlsx")
```

```python
sum = 0
bus_num = []
for i in range(0,len(bus_num_df["버스번호"])):
    bus_num.append(bus_num_df["버스번호"][i].split(" "))
```

```python
total = []

for i in range(0,len(bus_num_df["버스번호"])):
    sum = 0
    for j in range(0,len(bus_interval_df["버스번호"])):
        if bus_interval_df["버스번호"][j] in bus_num[i]:
            sum = sum + bus_interval_df["하루운행횟수"][j]
            print(sum)
    total.append(sum)        
```
